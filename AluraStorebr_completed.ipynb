```python
#Exibindo e calculando a quantidade de produtos vendidos por categoria.
vendas_por_categoria = loja.groupby('Categoria do Produto')['Produto'].count().reset_index()
print(vendas_por_categoria)
```

```python
#Criei uma função para calcular a quantidade de produtos vendidos por categoria e a categoria mais polular em vendas baseado no meu último código acima.
def calcular_vendas_por_categoria(csv_File):
  vendas_por_categoria = csv_File.groupby('Categoria do Produto')['Produto'].count().reset_index()
  print(vendas_por_categoria)

  #Usei um print a mais para dar mais espaço pra exibir a categoria mais popular
  print()

  #Categoria mais popular da loja
  categoria_produto_maisPopular = vendas_por_categoria['Produto'].idxmax()

  #Mostrar categoria mais vendida
  qnt_categoria_maisVendida = vendas_por_categoria['Produto'][categoria_produto_maisPopular]
  nome_categoria_maisVendida = vendas_por_categoria['Categoria do Produto'][categoria_produto_maisPopular]

  print(f"A categoria mais popular é: {nome_categoria_maisVendida}. O total de vendas para ela foi: {qnt_categoria_maisVendida}")
```

```python
#Calculando a quantidade de produtos vendidos por categoria da Loja1
print("Quantidade de produtos vendidos por categoria da Loja1: \n")
calcular_vendas_por_categoria(loja)
```

```python
#Esse outro aqui é bem similar ao código acima, mas soma todas as avalições para que possamos saber o total de avalições da loja
media_avaliacaoPorLoja = loja.groupby('Produto')['Avaliação da compra'].count().sum()
print(f'O número total de avaliações da loja1 foi de: {media_avaliacaoPorLoja}')
```

```python
'''
O challenge pede também a relação entre as avaliações dos clientes e os produtos
da loja, então criei essa função abaixo que exibe os produtos que receberam nota
máxima (5) e também os produtos que receberam nota mínima (1).
'''

def relacao_avaliacaoProduto_loja(csv_File):

  #Produtos com avaliação máxima (5)
  print("///////////////////////////////////////////////////////////////////////")
  print("Produtos com avalição 5:")
  print()

  produtos_avaliacao_5 = csv_File[csv_File['Avaliação da compra'] == 5]['Produto'].drop_duplicates().tolist()

  if produtos_avaliacao_5:

      for produto in produtos_avaliacao_5:
          print(produto)
  else:
      print("Nenhum produto com avaliação 5 encontrado.")

  print()
  #Produtos com avaliação máxima (1)
  print("///////////////////////////////////////////////////////////////////////")
  print("Produtos com avalição 1:")
  print()

  produtos_avaliacao_1 = csv_File[csv_File['Avaliação da compra'] == 1]['Produto'].drop_duplicates().tolist()

  if produtos_avaliacao_1:

      for produto in produtos_avaliacao_1:
          print(produto)
  else:
      print("Nenhum produto com avaliação 1 encontrado.")
```

```python
#Relação das avaliações com os produtos da Loja1
print("Relação das avaliações com os produtos da Loja1: \n")
relacao_avaliacaoProduto_loja(loja)
```

# Média de avalições por loja

```python
#Função que calcula a média de avaliações por loja
def media_avaliacoesLoja(csv_File):
  media_avaliacaoEstabelecimento = csv_File['Avaliação da compra'].mean().round(2)
  #print(f'A média de avaliações dessa loja é: {media_avaliacaoEstabelecimento}')
  print(media_avaliacaoEstabelecimento)
  return media_avaliacaoEstabelecimento
```

# 4. Produtos Mais e Menos Vendidos

```python
loja.head()
```

```python
#Criei uma função para calcular os produtos mais e menos vendidos da loja

def produtos_Mais_e_MenosVendidos(csv_File):

  # Contar a frequência de cada produto na coluna 'Produto'
  contagem_produtos = csv_File['Produto'].value_counts()

  # Exibir a contagem de cada produto
  #print("Contagem de vendas por produto:")
  #print(contagem_produtos)

  # Identificar o produto mais vendido
  produto_mais_vendido = contagem_produtos.idxmax()
  quantidade_mais_vendida = contagem_produtos.max()
  print(f"\nProduto mais vendido: {produto_mais_vendido} com {quantidade_mais_vendida} vendas")
  return produto_mais_vendido

  # Identificar o produto menos vendido
  produto_menos_vendido = contagem_produtos.idxmin()
  quantidade_menos_vendida = contagem_produtos.min()
  print(f"Produto menos vendido: {produto_menos_vendido} com {quantidade_menos_vendida} vendas")
  return produto_menos_vendido

  # Visualizar os 5 produtos mais vendidos
  print("\nTop 5 produtos mais vendidos:")
  print(contagem_produtos.head(5))

  # Visualizar os 5 produtos menos vendidos
  print("\nTop 5 produtos menos vendidos:")
  print(contagem_produtos.tail(5))
```

```python
print("Produtos mais e menos vendidos da Loja1: \n")
produtos_Mais_e_MenosVendidos(loja)

print("///////////////////////////////////////////////////////////////////////")
print()

print("Produtos mais e menos vendidos da Loja2: \n")
produtos_Mais_e_MenosVendidos(loja2)

print("///////////////////////////////////////////////////////////////////////")
print()

print("Produtos mais e menos vendidos da Loja3: \n")
produtos_Mais_e_MenosVendidos(loja3)

print("///////////////////////////////////////////////////////////////////////")
print()

print("Produtos mais e menos vendidos da Loja4: \n")
produtos_Mais_e_MenosVendidos(loja4)
```

```python
def calcular_freteMedio(csv_File):
  frete_Medio = csv_File['Frete'].mean()

  #Exibir o resultado
  print(f"O frete médio da loja é: R$ {frete_Medio:.2f}")
  return frete_Medio
```

```python
print("Frete médido para a Loja1")
calcular_freteMedio(loja)
print("/////////////////////////////////////")

print("\nFrete médido para a Loja2")
calcular_freteMedio(loja2)
print("/////////////////////////////////////")

print("\nFrete médido para a Loja3")
calcular_freteMedio(loja3)

print("/////////////////////////////////////")
print("\nFrete médido para a Loja4")
calcular_freteMedio(loja4)
```
